```python
import pandas as pd
from sqlalchemy import create_engine

import pymysql
pymysql.install_as_MySQLdb()
```

### Extract the CSVs into DataFrames

```python
crime_rate_csv = "Resources/2016_Annual_Crime_Data.csv"
crime_rate_df = pd.read_csv(crime_rate_csv)
crime_rate_df.head()
```

```python
graduation_rates_csv = "Resources/Austin_High_School_Graduation_Rates.csv"
graduation_rates_df = pd.read_csv(graduation_rates_csv)
graduation_rates_df.head()
```

### Transform the DataFrames

```python
crime_rate_df.head()
crime_rate_df.drop(columns=['GO Primary Key', 'Highest NIBRS/UCR Offense Description', 'Clearance Status', 'Clearance Date', 'GO Census Tract', 'GO X Coordinate', 'GO Y Coordinate'])
```

```python
new_crime_rate_df = crime_rate_df[['Council District', 'GO Highest Offense Desc', 'GO Report Date', 'GO Location']].copy()
new_crime_rate_df = new_crime_rate_df.rename(index=str, columns={"Council District": "council_district","GO Highest Offense Desc": "highest_offense_desc", "GO Report Date": "report_date", "GO Location": "location"})
new_crime_rate_df.set_index("council_district", inplace=False)
new_crime_rate_df.head()
```

```python
new_crime_rate_df.groupby(["council_district"]).count()
```

```python
# Dropped all unneccasry columns not relative to 2016
graduation_rates_df.head()
graduation_rates_df.drop(columns=['2015 Class Size', '2015 Graduated', '2015 Rate', '2014 Class Size', '2014 Graduated', '2014 Rate', '2013 Class Size', '2013 Graduated', '2013 Rate', '2012 Class Size', '2012 Graduated', '2012 Rate'])
```

```python
new_graduation_rates_df = graduation_rates_df[["School Type", "District", "School", "2016 Class Size", "2016 Graduated", "2016 Rate"]].copy()
new_graduation_rates_df = new_graduation_rates_df.rename(index=str, columns={"School Type": "school_type", "District": "district", "School": "school", "2016 Class Size": "2016_class_size", "2016 Graduated": "2016_graduated", "2016 Rate": "2016_rate"})
new_graduation_rates_df.head()
```

```python
new_graduation_rates_df.groupby(["school_type"]).count()
```

### Create Connection

```python
rds_connection_string = "root:password@localhost/etlscript_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
engine.table_names()
```

###  Load DataFrames into Database

```python
new_crime_rate_df.to_sql(name='crime_rate', con=engine, if_exists='append', index=False)
```

```python
new_graduation_rates_df.to_sql(name='graduation_rates', con=engine, if_exists='append', index=False)
```
